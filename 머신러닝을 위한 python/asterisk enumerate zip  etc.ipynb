List Comprehensions

````
result = []
for i in range(10) :
  result.append(i)
  
result
````

````
result = [ i for i in range(10)]
result
````

````
result = [ i for i in range(10) if i % 2 == 0 ]
result
````

````
# 1 dimension for loop 안에 for loop
word_1 = "hello"
word_2 = "world"
result = [i + j for i in word_1 for j in word_2]
result
````

````
case_1 =["A","B","C"]
case_2 = ["D","E","A"]
result = [i+j for i in case_1 for j in case_2]
result
````

````
result = [[a+b for a in case_1] for b in case_2]
result
````

````
result = [i+j for i in case_1 for j in case_2 if not(i==j)]
result
````

````
words = 'The quick brown fox jumps over the lazy dog'.split()
print(words)
## 2 dim
stuff = [[w.upper(), w.lower(), len(w)] for w in words]

for i in stuff:
    print(i)
  
````

# Zip 
> 두개의 list의 값을 병렬적으로 추출함

````
alist =[ "a1","a2","a3"]
blist= ["b1","b2","b3"]
for a,b in zip(alist, blist):
  print(a,b)
````

````
# vector 값 계산할 때!! 유용
a,b,c = zip((1,2,3),(10,20,30),(100,200,300))
print(a, b, c)
[sum(x) for x in zip(a, b, c)]
[sum(x) for x in zip((1,2,3),(10,20,30),(100,200,300))]
````

# Lambda
> 함수 이름 없이, 함수처럼 쓸 수있는 익명 함수

> 수학의 람다 대수에서 유래함

````
def f(x,y) :
  return x+y
print(f(1,4))
## mapreduce와 같이 쓸 때 효율적으로
f= lambda x,y : x+y
print(f(1,4))
````

````
f = lambda x:x**2
print(f(3))
````

## Map Reduce

````
ex= [1,2,3,4,5]
f = lambda x : x**2
print(list(map(f,ex)))
````

# Reduce function
> map function 과는 달리 list에 똑같은 함수를 적용해서 통합

> 옆으로 적용하면서 사용한다!

> 코드의 직관성이 떨어져서 lambda와 reduce는 사용하지 않지만 Legacy Library 나 머신러닝 코드에서는 여전히 사용중!

````
from functools import reduce
print(reduce(lambda x,y :x+y,[1,2,3,4,5]))
#1 +2 > 3+3 > 6+4 > 10 +5 => 15
````

````
def factorial(n) :
  return reduce( lambda x, y : x*y , range(1, n+1))

factorial(5)
````

# Asterisk
> 흔히 알고 있는 * 를 의미함

> 단순, 곱세브 제곱연산, 가변 인자 활용 등 다양한게 사용됨
### 다변인자



```
가변 인자
def asterisk_test(a, *args) :
  print(a, args)
  print(type(args))
 
키워드 인자
def asterisk_test(a, **kargs) :
  print(a, kargs)
  print(type(kargs))
```

# Asteriks - unpacking a container

- tuple ,dict 등  자료형에 들어가 있는 값을 unpacking
- 함수의 입력값 , zip 등에 유용하게 사용가능


````
# 1 은 a // 2,3,4,5 (나머지) 들은 *args 로 들어가라
def asterisk_test(a, *args) :
  print(a, args)
  print(type(args))
 

asterisk_test(1,2,3,4,5)

asterisk_test(1,3,4)

## 몇개가 들어올지 모를 때 한번에 묶어서 사용 할 수 있닫 좋은 방법
asterisk_test(1,(2,3,4,5,6))
# *args 에 (2,3,4,5,6)이 하나의 값으로 할당됨

## 차이를 비교해야 한다! tuple 꼴로 들어가기 때문에 
def asterisk_test2(a, *args) :
  print(a, args[0])
  print(type(args))
asterisk_test2(1,(2,3,4,5,6))

## 풀어서 던져주고싶을 때 
def asterisk_test3(a, args) : # (2,3,4,5,6) 으로 들어 왔다가 
  print(a, *args) # *args 로 효과로 인해 unpacking 이 된다
  print(type(args))
  
asterisk_test3(1,(2,3,4,5,6))

## 풀어서 던져주고싶을 때 2 -> 안먹네
def asterisk_test4(a, *args) : # (2,3,4,5,6) 으로 들어 왔다가 
  print(a, args) # *args 로 효과로 인해 unpacking 이 된다
  print(type(args))
  
asterisk_test4(1,*(2,3,4,5,6))

````

# Asterisk - unpacking a container

````
a ,b, c= ([1,2],[3,4],[5,6])
print(a,b,c)

data =([1,2],[3,4],[5,6])
print(*data)

def asterisk_test(a,b,c,d,) :
  print(a,b,c,d)
  
data ={"b" :1, "c":2 ,"d":3}
## ** 인자는 keyword 인자 -> dict type 뜯어서 들어가게 됨
asterisk_test(10, **data)
````

````
data =([1,2],[3,4],[5,6])
print(*data)


for data in zip(*([1,2],[3,4],[5,6])) :
  print(data)
  print(sum(data))
````

````
def asterisk_test(a,b,c,d,e=0) :
  print(a,b,c,d,e)
  
data = {"d" :1, "c":2 ,"b":3 ,"e":56}

asterisk_test(10, **data)
````

````

# dict type에 유용하게 사용함 따로 변수명 지정 안할 때 **kargs 활용 많이 함 
def asterisk_test(a, **kargs) :
  print(a, kargs)
  print(type(kargs))

asterisk_test(1, b=2, c=3, d=4, e=5, f= 6)
````
