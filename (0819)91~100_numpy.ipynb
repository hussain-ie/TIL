```
## 91 
# how to create a record array from a regular array ? 
import numpy as np

z = np.array([("HELLO" , 2.5 , 3) , 
             ("WORLD" , 3, 2)])

r = np.core.records.fromarrays(z.T , names='name, value1 , value2' , 
                              formats = 'S8 , f8, i8')

print(r)
```

```
## 93 
# Consider two arrays A and B of shape (8,3) and (2,2). 
# How to find rows of A that contain elements of each row of B 
# regardless of the order of the elements in B

a = np.random.randint(0, 5 , (8,3))
b = np.random.randint(0,5 , (2,2))
# print(np.shape(a[..., np.newaxis, np.newaxis]) )
print(a)
print(b)
# print(a[..., np.newaxis, np.newaxis] ) ## 더 잘게쪼개지는 역할을 하한다 
c = (a[..., np.newaxis, np.newaxis] ==b)  ## 더 잘게 쪼개진것에서 2x2 행렬중에 있는 값이 있으면 TRUE 
print(a[..., np.newaxis, np.newaxis][1:3,1:3,:,:])
print(c[1:3,1:3,:,:] )
print(np.shape(c))
print(c.any((3,1)))
print("\n" ,c.any((3,1)).all(1))
print( np.where(c.any((3,1)).all(1)) )
rows = np.where(c.any((3,1)).all(1))[0]
print(rows)
```

```
# 94
## considering a 10x3 matrix , extract rows with unequal values

z= np.random.randint(0 ,5 , (10, 3 ))
print(z[:,1:] ,"\n")
print(z[:,:-1])
e = np.all(z[:,1:] == z[:,:-1] ,axis = 1)
print(e, "\n" , ~e)
u=z[~e]
print(  u)
u = z[z.max(axis=1) != z.min(axis=1) , : ]
print("\n" , u)
```

```
# 96
## given a two dimensional array , how to extract unique rows
z = np.random.randint(0, 2, (6,3))
t = np.ascontiguousarray(z).view(np.dtype( (np.void, z.dtype.itemsize * z. shape[1]) ) ) 
_, idx = np.unique(t , return_index =True)
print(z)
print(idx , "\n")

uZ = z[idx]
print(uZ)
```

```
# 97
## Considering 2 vectors A & B, 
## write the einsum equivalent of inner, outer, sum, and mul function (★★★)

a= np.random.uniform(0, 1, 10)
b= np.random.uniform(0, 1, 10)

np.einsum("i->", a) # np.sum
np.einsum("i,i->i"  , a,b) # a*b
np.einsum("i,i" , a,b) # np.inner
np.einsum("i,j->ij" , a,b) #np.outer
```

```
# 98
# Considering a path described by two vectors (X,Y), 
# how to sample it using equidistant samples (★★★)?

phi = np.arange(0, 10*np.pi , 0.1)
a =1 
x = a*phi*np.cos(phi)
y= a*phi*np.sin(phi)

dr = (np.diff(x)**2  + np.diff(y)**2)**.5
r = np.zeros_like(x)
print(np.shape(r) == np.shape(x))
r[1:] = np.cumsum(dr)
print(r)
r_int = np.linspace(0, r.max() , 200)
x_int = np.interp( r_int , r, x)
print(x_int)
y_int = np.interp( r_int ,r , y )
```

```
# 100
# Compute bootstrapped 95% confidence intervals for the mean of a 1D array X 
# (i.e., resample the elements of an array with replacement N times, compute the mean of each sample, and then compute percentiles over the means).


x  = np.random.randn(100)
n = 1000 
idx = np.random.randint( 0 ,x.size , (n, x.size)) 
means = x[idx].mean(axis=1)
print( np.shape( means) )
confint = np.percentile( means, [2.5 , 97.5 ])
print(confint)








```
