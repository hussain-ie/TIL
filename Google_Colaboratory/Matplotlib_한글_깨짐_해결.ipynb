* # ticks 설정
plot 이나 chart 에서 축상의 위치가 표시 지점을 **tick** 이라 한다.   
이 tick은 써진 숫자 혹은 글자를  tick label 이라고 한다.   
tick의 위치나 tick label 은 matplotlib 에서 자동으로 정해주지만, 만약 수동으로 설정하고 싶다면 xticks 명령이나 yticks 명령을 사용한다.  
tick label 문자열에는 $ $ 사이에 Latex 를 이용해 숫자 문자식도 가능하다.

````
import numpy as np
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.plot(X,C)
plt.xticks([-np.pi , -np.pi/2, 0, np.pi /2 , np.pi ])
plt.yticks([-1, 0, 1])
plt.show()
````

````
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.plot(X,C)
plt.xticks([-np.pi , -np.pi/2 , 0 , np.pi/2 , np.pi] , [r'$-\pi$' , r'$-\pi/2$' ,r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1,0,1],["LOW", 'ZERO', " HIGH"])
plt.show()
````

* # hold 명령
하나의 plot 명령이 아니라 복수의 plot 명령을 하나의 그림에서 겹쳐서 그릴 수 있다.   
matplotlib 1.5 까지는 hold(True) 명령을 이용하면  기존의 그림 위에 겹쳐 그리도록 하였다.   
겹치기를 종료하는 것은 hold(False)  
matplotlib 2.0 부터는 자동으로 된다.

````
plt.plot([1, 4, 9, 16], c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.hold(False)
plt.plot([9, 16, 4, 1], c="k", lw=3, ls=":", marker="s", ms=10, mec="m", mew=5, mfc="c")

plt.show()
````

* # Figure 

모든 그림은 Figure 객체,  Matplotlib.figure.Figure 클래스 객체에 포함되어  있다.  

내부 플롯(inline plot)이 아닌 경우에는 하나의 Figure는 하나의 아이디 숫자와 윈도우(Window)를 가진다.    

주피터 노트북에서는 윈도우 객체가 생성되지 않지만 파이썬을 독립 실행하는 경우에는 하나의 Figure당 하나의 윈도우를 별도로 가진다.  

원래 Figure를 생성하려면 figure 명령을 사용하여 그 반환값으로 Figure 객체를 얻어야 한다.   

그러나 일반적인 plot 명령 등을 실행하면 자동으로 Figure를 생성해주기 때문에 일반적으로는 figure 명령을 잘 사용하지 않는다  

figure 명령을 명시적으로 사용하는 경우는 여러개의 윈도우를 동시에 띄워야 하거나(line plot이 아닌 경우),   

Jupyter 노트북 등에서(line plot의 경우) 그림의 크기를 설정하고 싶을 때이다.   
그림의 크기는 figsize 인수로 설정한다.

[Reference](http://Matplotlib.org/api/figure_api.html#Matplotlib.figure.Figure)


````
np.random.seed(0)
f1 = plt.figure(figsize=(10,2))
plt.plot(np.random.randn(100))
plt.show()
````

* 만약 명시적으로 figure 명령을 사용하지 않는 경우  figure 객체를 얻으려면 gcf 명령을 사용한다.

````
f1 = plt.figure(1)
plt.plot([1,2,3,4],"ro:")
f2 = plt.gcf()
print(f1, id(f1))
print(f2, id(f2))
plt.show()
````

* # Axes 객체와 subplot 명령
하나의 윈도우 즉 **figure** 안에 여러개의 plot 을 배열 형태로 보이고 싶을 때가 이다.  
Figure 안에 있는 각각의 plot 은 Axes 라고 불리는 객체에 속한다.  [참고](http://Matplotlib.org/api/axes_api.html#Matplotlib.axes.Axes)  
Figure 안에 Axes 를  생성하려면 원래는 **subplot** 명령을 사용하여 명시적으로 Axes 객체를 얻어야 한다.  
하지만 plot 명령을 바로 사용해도 자동으로 Axed 를 생성해준다.  
**subplot** 명령은 grid 형태의 Axes 객체를 생성하는데. Figrue 가 matrix 이고 Axes 가 행렬의 원소 같은 개념이 된다.  
ex 위 아래 2개의 plot  2 x 1 행렬에다가 넣어주면 된다.   
* 여기에서 숫자 인덱싱은 파이썬이 아닌 Matlab 관행을 따르기 때문에 첫번째 플롯을 가리키는 숫자가 0이 아니라 1임에 주의하라.



```
subplot(2, 1, 1)
# 여기에서 윗부분에 그릴 플롯 명령 실행
subplot(2, 1, 2)
# 여기에서 아랫부분에 그릴 플롯 명령 실행
```



````
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
print(ax1)

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
print(ax2)

plt.tight_layout()
plt.show()
````

````
np.random.seed(0)

plt.subplot(221)
plt.plot(np.random.rand(5))
plt.title("axes 1")

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title("axes 2")

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title("axes 3")

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title("axes 4")

plt.tight_layout()
plt.show()
````

### subplots 명령으로 복수의 Axes 객체를 동시에 생성할 수도 있다. 이때는 2차원 ndarray 형태로 Axes 객체가 반환된다.

````
fig, axes = plt.subplots(2, 2)

np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 0].set_xlabel("hi")
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[0, 1].set_ylabel("bye")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.tight_layout()
plt.show()
````

* #  Axis 객체와 축¶
하나의 Axes 객체는 두 개 이상의 Axis 객체를 가진다. Axis 객체는 플롯의 가로축이나 세로축을 나타내는 객체이다. [참고](https://matplotlib.org/api/axis_api.html) 여러가지 plot 을 하나의 Axes 객체에 표시 할 때 y 값의 크기가 다랄서 표시하기 힘든 경우가 있다.  
이 때는 다음처럼 **twinx**명령으로 대해 복수의 y 축을 가진 plot을 만들 수 있다.  
twinx 명령은 x 축을 공유하는 새로운 Axes 객체를 만든다.  



````
fig , ax0 = plt.subplots()
ax1 = ax0.twinx()

ax0.plot([10,5,9,2,4], "r-", label="y0")
ax0.set_ylabel("y0")
ax0.grid(False)
ax0.legend(loc=3)
ax1.plot([100, 200, 220, 180, 120], 'g:', label="y1")
ax1.set_ylabel("y1")
ax1.grid(False)
ax1.legend(loc=2)
plt.show()
````
